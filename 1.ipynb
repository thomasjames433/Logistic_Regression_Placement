```python
import pandas as pd
import numpy as np
df=pd.read_csv("placement.csv")
```

```python
df["PlacementStatus"]=(df["PlacementStatus"]=="Placed").astype(int)
df["ExtracurricularActivities"]=(df["ExtracurricularActivities"]=="Yes").astype(int)
df["PlacementTraining"]=(df["PlacementTraining"]=="Yes").astype(int)
```

```python
X=df[[
    "CGPA",
    "Internships",
    "Projects",
    "Workshops/Certifications",
    "AptitudeTestScore",
    "SoftSkillsRating",
    "ExtracurricularActivities",
    "PlacementTraining",
    "SSC_Marks",
    "HSC_Marks"
]].values
```

```python
y=df["PlacementStatus"].values
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X=scaler.fit_transform(X)
```

```python
from logistic_regression import gradient_descent,predict

theta= gradient_descent(X,y)
print(theta)
```

```python
y_pred=predict(X,theta)
```

```python
accuracy=np.mean(y_pred==y)
print(accuracy)
```
